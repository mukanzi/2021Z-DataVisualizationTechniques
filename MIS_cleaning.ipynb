#Cleaning of MIS data to produce alumni contacts.

The below code is to clean MIS data so as to produce alumni contacts, name, phone number and email.

## Importing the data

```python
import pandas as pd
df = pd.read_csv("/content/FINAL Flat Data Report.csv")
df.head ()
```

```python
df_alumni = df[df['Status']=='Alumni']
df_alumni.head()
```

```python
df_alumni.shape
```

```python
df_alumni.drop_duplicates()
df_alumni.shape
```

```python
df_alumni.columns
```

```python
df_alumni_contacts = df_alumni[['Case Number','Status','Name', 'Most Recent Scholar Email 1', 'Most Recent Scholar Email 2','Most Recent Scholar Inst Email']]
df_alumni_contacts.head()
```

```python
df_alumni_contacts.shape
```

```python
df_cleaned = df_alumni_contacts.dropna(subset=['Most Recent Scholar Email 1', 'Most Recent Scholar Email 2', 'Most Recent Scholar Inst Email'], how='all')
df_cleaned.shape
```

```python
df_alumni_contacts.to_excel("Alumni_contacts_not_clean.xlsx")
df_cleaned.to_csv("Cleaned_alumni_Contacts.xlsx")
```
